```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_matkul = [
    ("Fungsional", 87.5, 48),
    ("Jarkom", 91, 52),
    ("Mobile", 79.7, 40),
    ("Pirdas", 92.5, 49),
    ("Web", 89, 55)
]

dataset = data_matkul
```

```python
subjek = [item[0] for item in dataset]
nilai = [item[1] for item in dataset]
jam = [item[2] for item in dataset]
```

```python
# Memakai Plotting menggunakan pyplot

plt.figure(figsize=(16,8))
plt.title('Data Matkul', fontsize=30)
plt.plot(subjek, nilai, marker='o', linestyle='-', color='blue', label='Nilai')
plt.plot(subjek, jam, marker='x', linestyle='-', color='red', label='Jam Belajar')
plt.xlabel('Mata Kuliah', fontsize=20)
plt.ylabel('Nilai / Jam Belajar', fontsize=20)
plt.xticks(rotation=45, fontsize=15)
plt.yticks(fontsize=15)
plt.legend(fontsize=15)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plotting menggunakan Scatter Plot

plt.figure(figsize=(16,8))
plt.title('Data Matkul', fontsize=30)
plt.scatter(subjek, nilai, color='blue', label='Nilai', s=100)
plt.scatter(subjek, jam, color='red', label='Jam Belajar', s=100)
plt.xlabel('Mata Kuliah', fontsize=20)
plt.ylabel('Nilai / Jam Belajar', fontsize=20)
plt.xticks(rotation=45, fontsize=15)
plt.yticks(fontsize=15)
plt.legend(fontsize=15)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Visualisasi data menggunakan Pie Chart

# Membuat pie chart untuk Nilai
plt.figure(figsize=(16,8))
plt.pie(nilai, labels=subjek, autopct='%1.1f%%', startangle=140)
plt.title('Distribusi Nilai per Mata Kuliah', fontsize=20)
plt.show()

# Membuat pie chart untuk Jam Belajar
plt.figure(figsize=(8,8))
plt.pie(jam, labels=subjek, autopct='%1.1f%%', startangle=140)
plt.title('Distribusi Jam Belajar per Mata Kuliah', fontsize=20)
plt.show()
```

```python
# Subplot dalam matplotlib

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))

ax1.pie(nilai, labels=subjek, autopct='%1.1f%%', startangle=140, colors=plt.cm.viridis(np.linspace(0, 1, len(subjek))))
ax1.set_title('Distribusi Nilai per Mata Kuliah', fontsize=20)

ax2.pie(jam, labels=subjek, autopct='%1.1f%%', startangle=140, colors=plt.cm.plasma(np.linspace(0, 1, len(subjek))))
ax2.set_title('Distribusi Jam Belajar per Mata Kuliah', fontsize=20)

plt.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(24, 8))

ax1.pie(nilai, labels=subjek, autopct='%1.1f%%', startangle=140, colors=plt.cm.viridis(np.linspace(0, 1, len(subjek))))
ax1.set_title('Distribusi Nilai per Mata Kuliah', fontsize=20)
ax2.pie(jam, labels=subjek, autopct='%1.1f%%', startangle=140, colors=plt.cm.plasma(np.linspace(0, 1, len(subjek))))
ax2.set_title('Distribusi Jam Belajar per Mata Kuliah', fontsize=20)

width = 0.35 
x = np.arange(len(subjek))

rects1 = ax3.bar(x - width/2, nilai, width, label='Nilai', color='blue')
rects2 = ax3.bar(x + width/2, jam, width, label='Jam Belajar', color='red')

ax3.set_xlabel('Mata Kuliah', fontsize=20)
ax3.set_ylabel('Nilai / Jam Belajar', fontsize=20)
ax3.set_title('Perbandingan Nilai dan Jam Belajar', fontsize=20)
ax3.set_xticks(x)
ax3.set_xticklabels(subjek, rotation=45, fontsize=15)
ax3.legend()

plt.tight_layout()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(16, 8))

bar_width = 0.4
x = np.arange(len(subjek))
bars = ax1.bar(x - bar_width/2, jam, bar_width, label='Jam Belajar', color='red')

ax2 = ax1.twinx()
line, = ax2.plot(x + bar_width/2, nilai, label='Nilai', color='blue', marker='o')

ax1.set_xlabel('Mata Kuliah', fontsize=20)
ax1.set_ylabel('Jam Belajar', fontsize=20, color='red')
ax2.set_ylabel('Nilai', fontsize=20, color='blue')
ax1.set_title('Perbandingan Nilai dan Jam Belajar per Mata Kuliah', fontsize=24)
ax1.set_xticks(x)
ax1.set_xticklabels(subjek, rotation=45, fontsize=15)

bars_legend = ax1.legend(loc='upper left', fontsize=15)
line_legend = ax2.legend(loc='upper right', fontsize=15)

plt.grid(True)
plt.tight_layout()
plt.show()
```
