```python
import pandas as pd
```

```python
dataset = pd.read_csv('Dataset Harga Buah dan Sayur.csv')
dataset.head(10)
```

```python
dataset['date'] = pd.to_datetime(dataset['date'])
dataset['year'] = dataset['date'].dt.year
```

```python
def RataRata(df, Gkolom, groupby_cols):
    result = df.groupby(groupby_cols)[Gkolom].mean().reset_index()
    result['Rata-Rata'] = result[Gkolom].round(2)
    return result.drop(columns=[Gkolom])

def Minimum(df, Gkolom, groupby_cols):
    result = df.groupby(groupby_cols)[Gkolom].min().reset_index()
    result['Harga Terendah'] = result[Gkolom].round(2)
    return result.drop(columns=[Gkolom])

def Maksimal(df, Gkolom, groupby_cols):
    result = df.groupby(groupby_cols)[Gkolom].max().reset_index()
    result['Harga Tertinggi'] = result[Gkolom].round(2)
    return result.drop(columns=[Gkolom])
```

```python
def hitung_statistik_tahunan(dataset):
    statistik = dataset.groupby(['year', 'category', 'item']).agg({
        'price': ['mean', 'min', 'max']
    }).reset_index()
    statistik.columns = ['Tahun', 'Kategori', 'Item', 'Rata-Rata', 'Harga Terendah', 'Harga Tertinggi']
    
    statistik['Rata-Rata'] = statistik['Rata-Rata'].round(2)
    statistik['Harga Terendah'] = statistik['Harga Terendah'].round(2)
    statistik['Harga Tertinggi'] = statistik['Harga Tertinggi'].round(2)
    
    return statistik

statistik_tahunan = hitung_statistik_tahunan(dataset)
```

```python
filter_harga = dataset[(dataset['price'] >= 1.50) & (dataset['price'] <= 2.35)]
```

```python
averange = RataRata(dataset, 'price', ['year', 'category', 'item'])
Minimum = Minimum(dataset, 'price', ['year', 'category', 'item'])
Maximum = Maksimal(dataset, 'price', ['year', 'category', 'item'])
```
