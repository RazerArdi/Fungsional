## 3.3 Perulangan Dalam Python

Pada python, kita bisa melakukan perulangan dengan beberapa cara diantaranya:

### **3.3.1 Perulangan For**


<p align="justify">
Perulangan for pada python adalah perintah yang digunakan <b> untuk melakukan iterasi</b> dari sebuah nilai <b> sequence </b> atau data koleksi seperti List, Tuple, String, range dan lainnya. For pada python memiliki perilaku yang berbeda dengan for pada kebanyakan bahasa pemrograman yang lain, karena pada python ia sangat berkaitan dengan data sequence atau data kolektif. Mungkin kalau dibandingkan dengan bahasa lain, for pada python lebih dikenal sebagai foreach. Perhatikan contoh berikut:


```python
# for loop pada sequence type

friendList = ["Sutadi", "Andri", "Boy", "Bagus", "Widodo"]
for x in friendList:
  print(x)
```

```python
# for loop pada string

for x in "Zaid":
  print(x)
```

```python
# break keyword

for x in friendList:
  print(x)
  if x == "Bagus":
    print("Next? kita cut off dulu yaa")
    break
```

```python
# continue keyword

for x in friendList:
  if x == "Boy":
    print("Yang satu ini kita skip yaa")
    continue
  print(x)
```

```python
# range dalam for loop

for x in range(6):
  print("Astagfirullah")
```

```python
# nested for

for x in range(3):
  for y in range(5):
    print(y, end=" ") # mencegah baris baru
  print() # membuat baris baru
```

```python
# range 2 argumen
for i in range(2, 6):
    print(i)
```

```python
# range 3 argumen
for i in range(1, 10, 2):
    print(i)
```

### 3.3.3 Fungsi Rekursif

<p align="justify">
Fungsi rekursif adalah fungsi yang melakukan perulangan dengan mengacu pada dirinya sendiri. Dalam paradigma pemrograman fungsional, rekursi adalah cara penting untuk melakukan iterasi data, menggantikan penggunaan loop konvensional.

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)


print(factorial(5))
```

<p align="justify">
Rekursi dalam paradigma fungsional mempromosikan penggunaan fungsi murni dan menghindari perubahan <b>keadaan (state)</b> yang tidak perlu. Fungsi murni adalah fungsi yang hasilnya hanya bergantung pada inputnya dan tidak mengubah variabel di luar fungsi tersebut. Terkait hal ini akan kita bahas lebih detil pada modul selanjutnya.


## 3.4 Function Dalam Python

<p align="justify"> Jika pada paradigma pemrograman sebelumnya (OOP-Java) kalian lebih familiar dengan istilah method, maka pada pemrograman fungsional dengan python kali ini kita akan banyak menggunakan function (ee, jangan salah, di python juga ada method kok, cek cek <a href="https://medium.com/@rismitawahyu/function-and-methods-in-python-fungsi-dan-metode-di-python-757204314517">disini</a> untuk tau perbedaan keduanya). untuk tau perbedaan keduanya)


<p align="justify"> Python sendiri memiliki banyak built-in function, diantaranya yang sering kita gunakan, antara lain print() untuk mencetak objek, len() berfungsi mengembalikan (return) panjang suatu objek, type() berfungsi mengembalikan (return) tipe data dari suatu variabel, dan masih banyak lagi yang lainnya. Sedangkan untuk membuat suatu function, perlu didefinisikan dengan keyword “def”. Untuk lebih memahami mengenai function pada python, kalian dapat menjalankan program dibawah ini:

```python
def firstExample():
    print("Hello it's me")

firstExample()

# Output : Hello it's me

def secondExample(name, hobbies):
    print("Hi, iam " + name + " and my hobby is " + hobbies)

secondExample("Widodo", "Fishing")

# Output : Hi, iam Widodo and my hobby is Fishing
```

<p align="justify"> Fungsi-fungsi diatas adalah deklarasi fungsi standar pada python. Meski kita sudah membangun program kita dalam satuan fungsi-fungsi, bukan berarti kita sudah mengimplementasikan paradigma fungsional. Seperti halnya fungsi-fungsi diatas, fungsi-fungsi tersebut belum memenuhi paradigma fungsional.

<p align="justify"> Seperti apakah fungsi dalam paradigma fungsional itu? Mari kita cek kembali materi topik 1 tentang definisi fungsi dalam paradigma fungsional:

 - Program dalam pemrograman fungsional dapat dianggap sebagai kumpulan **fungsi yang menerima input dan menghasilkan output** tanpa mempengaruhi keadaan (stateless) dan tidak memiliki efek samping apapun (pure function).

<p align="justify"> Kita bisa memperbaiki dan menyusun kode dengan pendekatan lebih fungsional untuk mencapai manfaat yang ditawarkan oleh paradigma fungsional. Bagaimana caranya? Pastikan fungsi yang kita bangun memenuhi standar berikut:

1. Input: tambahkan parameter sebagai input data-data yang dibutuhkan oleh 2. fungsi.
2. Output: tambahkan return value sebagai hasil keluaran dari fungsi.
3. Pure: tidak memiliki efek samping berarti fungsi tidak mengakses atau mengubah variabel global atau data-data lain di luar fungsi.

Dengan demikian, kita dapat menghasilkan program yang lebih elegan, mudah dipahami, dan mudah dikelola. Mari kita perbaiki dan susun ulang kode fungsi sebelumnya dengan pendekatan fungsional sebagai berikut: </p>



```python
# deklarasi fungsi menggunakan parameter (reusability)
def secondExample(name, hobbies):
    # Menambahkan return sebagai output
    return ("Hi, iam " + name + " and my hobby is " + hobbies)

# Assign function ke dalam variabel
callback = secondExample("Widodo", "Fishing")


print(callback)
```

```python
def secondExample(name, hobbies):
    return ("Hi, iam " + name + " and my hobby is " + hobbies)

callback = secondExample("Widodo", "Fishing")

del secondExample

print(callback)
```

### Codelab 3

1. Buatlah sebuah kamus (dictionary) yang menyimpan data nilai untuk 5 mahasiswa dengan 3 mata kuliah berbeda.
2. Buatlah fungsi untuk menghitung rata-rata nilai setiap mahasiswa.
3. Buatlah fungsi untuk menghitung rata-rata nilai seluruh mahasiswa.


```python
# Data nilai mahasiswa dengan 3 mata kuliah
nilai_mahasiswa = {
    "Bayu": {"Matematika": 85, "Fisika": 78, "Kimia": 92},
    "Uyyab": {"Matematika": 70, "Fisika": 88, "Kimia": 75},
    "Ardi": {"Matematika": 95, "Fisika": 90, "Kimia": 85},
    "Ardiyansyah": {"Matematika": 80, "Fisika": 85, "Kimia": 78},
    "Ubay": {"Matematika": 88, "Fisika": 72, "Kimia": 91}
}

# TO_DO 2: Function untuk menghitung rata-rata nilai setiap mahasiswa
def each_average(nilai_mahasiswa):
    rata_rata_mahasiswa = {}

    for mahasiswa, nilai in nilai_mahasiswa.items():
        total_nilai = sum(nilai.values())  
        rata_rata = total_nilai / len(nilai)
        rata_rata_mahasiswa[mahasiswa] = rata_rata 

    return rata_rata_mahasiswa

# TO_DO 3: Function untuk menghitung rata-rata nilai seluruh mahasiswa
def all_average(nilai_mahasiswa):
    total_semua_nilai = 0
    total_semua_mata_kuliah = 0

    for nilai in nilai_mahasiswa.values():
        total_semua_nilai += sum(nilai.values()) 
        total_semua_mata_kuliah += len(nilai)

    rata_rata_seluruh_mahasiswa = total_semua_nilai / total_semua_mata_kuliah
    return rata_rata_seluruh_mahasiswa

# Memanggil fungsi dan menampilkan hasil

# Rata-rata nilai setiap mahasiswa
rata_rata_per_mahasiswa = each_average(nilai_mahasiswa)
print("Rata-rata nilai tiap mahasiswa:")
for mahasiswa, rata_rata in rata_rata_per_mahasiswa.items():
    print(f"{mahasiswa}: {rata_rata:.2f}")

# Rata-rata nilai seluruh mahasiswa
rata_rata_semua_mahasiswa = all_average(nilai_mahasiswa)
print(f"\nRata-rata nilai seluruh mahasiswa: {rata_rata_semua_mahasiswa:.2f}")
```

```python
accounts = {}  # Key: NIM, Value: {"password": pass, "role": role}
personnel_data = []  # List of Dictionaries: {"nim": str, "name": str, "rank": str, "unit_id": int}
unit_data = []  #  {"unit_id": int, "unit_name": str}
mission_data = []  #  (mission_id, mission_name, location, [involved_personnel])

def register():
    nim = input("Enter your NIM: ")
    if nim in accounts:
        print("NIM already registered!")
        return
    password = input("Enter your password: ")
    role = input("Enter your role (admin/user): ")
    accounts[nim] = {"password": password, "role": role}
    print("Account successfully registered!")
    if role == "user":
        add_profile(nim)

def add_profile(nim):
    name = input("Enter your name: ")
    rank = input("Enter your rank: ")
    try:
        unit_id = int(input("Enter your Unit ID (0 if unknown): "))
    except ValueError:
        print("Invalid Unit ID. Profile not saved.")
        return
    if unit_id != 0 and not any(u["unit_id"] == unit_id for u in unit_data):
        print(f"Unit ID {unit_id} not found. Profile not saved.")
        return
        
    existing_profile = next((p for p in personnel_data if p["nim"] == nim), None)
    if existing_profile:
        print(f"Profile for NIM {nim} already exists.")
    else:
        personnel_data.append({"nim": nim, "name": name, "rank": rank, "unit_id": unit_id})
        print(f"Profile for NIM {nim} has been added.")

def login():
    nim = input("Enter your NIM: ")
    if nim not in accounts:
        print("NIM not found. Please register first.")
        return None
    password = input("Enter your password: ")
    if accounts[nim]["password"] == password:
        print(f"Welcome, {nim}!")
        return nim
    else:
        print("Incorrect password.")
        return None

def view_profile(nim):
    profile = next((p for p in personnel_data if p["nim"] == nim), None)
    if profile:
        unit = next((u for u in unit_data if u["unit_id"] == profile["unit_id"]), {"unit_name": "Unknown"})
        print(f"Name: {profile['name']}, Rank: {profile['rank']}, Unit: {unit['unit_name']}")
    else:
        print(f"No profile found for NIM {nim}. Please update your profile.")

def update_profile(nim):
    profile = next((p for p in personnel_data if p["nim"] == nim), None)
    if profile:
        print(f"Current Profile: Name={profile['name']}, Rank={profile['rank']}, Unit ID={profile['unit_id']}")
        profile['name'] = input(f"Enter new name (leave blank to keep {profile['name']}): ") or profile['name']
        profile['rank'] = input(f"Enter new rank (leave blank to keep {profile['rank']}): ") or profile['rank']
        try:
            new_unit_id = input(f"Enter new Unit ID (leave blank to keep {profile['unit_id']}): ")
            if new_unit_id:
                new_unit_id = int(new_unit_id)
                if new_unit_id != 0 and not any(u["unit_id"] == new_unit_id for u in unit_data):
                    print(f"Unit ID {new_unit_id} not found. Update cancelled.")
                else:
                    profile['unit_id'] = new_unit_id
        except ValueError:
            print("Invalid Unit ID. Update cancelled.")
        print("Profile updated successfully.")
    else:
        print(f"No profile found for NIM {nim}. Please add your profile first.")

def delete_profile(nim):
    global personnel_data
    personnel_data = [p for p in personnel_data if p["nim"] != nim]
    del accounts[nim]
    print(f"Profile and account for NIM {nim} deleted successfully.")

def add_personnel():
    nim = input("Enter NIM of the new personnel: ")
    if nim in accounts:
        print("NIM already exists in accounts.")
        return
    register() 

def add_unit():
    try:
        unit_id = int(input("Enter Unit ID: "))
        unit_name = input("Enter Unit Name: ")
        if any(u["unit_id"] == unit_id for u in unit_data):
            print(f"Unit ID {unit_id} already exists.")
        else:
            unit_data.append({"unit_id": unit_id, "unit_name": unit_name})
            print(f"Unit {unit_name} added successfully.")
    except ValueError:
        print("Invalid Unit ID. Operation cancelled.")

def add_mission():
    try:
        mission_id = int(input("Enter Mission ID: "))
        mission_name = input("Enter Mission Name: ")
        location = input("Enter Mission Location: ")
        involved_personnel = input("Enter involved personnel NIMs (comma-separated): ").split(',')
        mission_data.append((mission_id, mission_name, location, involved_personnel))
        print(f"Mission {mission_name} added successfully.")
    except ValueError:
        print("Invalid Mission ID. Operation cancelled.")

def delete_personnel():
    nim = input("Enter NIM of the personnel to delete: ")
    if nim in accounts:
        delete_profile(nim)
    else:
        print(f"No personnel found with NIM {nim}.")

def delete_unit():
    try:
        unit_id = int(input("Enter Unit ID to delete: "))
        global unit_data
        unit_data = [u for u in unit_data if u["unit_id"] != unit_id]
        print(f"Unit ID {unit_id} deleted successfully.")
    except ValueError:
        print("Invalid Unit ID. Operation cancelled.")

def delete_mission():
    try:
        mission_id = int(input("Enter Mission ID to delete: "))
        global mission_data
        mission_data = [m for m in mission_data if m[0] != mission_id]
        print(f"Mission ID {mission_id} deleted successfully.")
    except ValueError:
        print("Invalid Mission ID. Operation cancelled.")

def view_all_personnel():
    print("All Personnel:")
    for personnel in personnel_data:
        print(personnel)

def view_all_units():
    print("All Units:")
    for unit in unit_data:
        print(unit)

def view_all_missions():
    print("All Missions:")
    for mission in mission_data:
        print(mission)

def admin_menu():
    while True:
        print("\nAdmin Menu:")
        print("1. Add Personnel")
        print("2. Add Unit")
        print("3. Add Mission")
        print("4. Delete Personnel")
        print("5. Delete Unit")
        print("6. Delete Mission")
        print("7. View All Personnel")
        print("8. View All Units")
        print("9. View All Missions")
        print("0. Logout")
        choice = input("Choose an option: ")
        if choice == "1":
            add_personnel()
        elif choice == "2":
            add_unit()
        elif choice == "3":
            add_mission()
        elif choice == "4":
            delete_personnel()
        elif choice == "5":
            delete_unit()
        elif choice == "6":
            delete_mission()
        elif choice == "7":
            view_all_personnel()
        elif choice == "8":
            view_all_units()
        elif choice == "9":
            view_all_missions()
        elif choice == "0":
            break
        else:
            print("Invalid choice.")

def user_menu(nim):
    while True:
        print("\nUser Menu:")
        print("1. View Profile")
        print("2. Update Profile")
        print("3. Delete Profile")
        print("0. Logout")
        choice = input("Choose an option: ")
        if choice == "1":
            view_profile(nim)
        elif choice == "2":
            update_profile(nim)
        elif choice == "3":
            confirm = input("Are you sure you want to delete your profile? (yes/no): ")
            if confirm.lower() == 'yes':
                delete_profile(nim)
                break
        elif choice == "0":
            break
        else:
            print("Invalid choice.")

def main():
    while True:
        print("\nMain Menu:")
        print("1. Register")
        print("2. Login")
        print("0. Exit")
        choice = input("Choose an option: ")
        if choice == "1":
            register()
        elif choice == "2":
            nim = login()
            if nim:
                if accounts[nim]["role"] == "admin":
                    admin_menu()
                else:
                    user_menu(nim)
        elif choice == "0":
            break
        else:
            print("Invalid choice.")

main()
```
