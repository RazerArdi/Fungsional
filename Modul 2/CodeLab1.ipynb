```python
nilai = [
    {'matkul': 'Alpro', 'nilai': 80},
    {'matkul': 'Strukdat', 'nilai': 90},
    {'matkul': 'ADB', 'nilai': 75}
]

'''
def tambah_nilai(nama_matkul, jumlah_nilai):
    global nilai
    for item in nilai:
        if item['matkul'] == nama_matkul:
            item['nilai'] += jumlah_nilai
            print(f"Nilai {nama_matkul} ditambahkan {jumlah_nilai}. \nTotal nilai {nama_matkul} saat ini {item['nilai']}")
            break
    else:
        print(f"Mata kuliah {nama_matkul} tidak ditemukan dalam daftar!")

# Cek nilai awal
print("Nilai awal: ",nilai)

# Menambahkan nilai ke dalam daftar
tambah_nilai('Mobile', 15)

# Cek nilai saat ini
print("Nilai update:",nilai)
'''
# a. Fungsi tambah_nilai (pure function)
def tambah_nilai(data_nilai, nama_matkul, jumlah_nilai):
  """Menambahkan nilai ke mata kuliah tertentu dalam data nilai.

  Args:
    data_nilai: List of dictionaries, dengan setiap dictionary berisi 'matkul' dan 'nilai'.
    nama_matkul: String, nama mata kuliah yang akan ditambah nilainya.
    jumlah_nilai: Integer, nilai yang akan ditambahkan.

  Returns:
    List of dictionaries: Data nilai yang sudah diupdate.
  """
  hasil = []
  for item in data_nilai:
    new_item = item.copy()
    if item['matkul'] == nama_matkul:
      new_item['nilai'] += jumlah_nilai
    hasil.append(new_item)
  return hasil

# b. Fungsi generator cetak_nilai
# Fungsi generator cetak_nilai
def cetak_nilai(data_nilai, nim):
  """Generator untuk mencetak nilai.

  Args:
    data_nilai: List of dictionaries, dengan setiap dictionary berisi 'matkul' dan 'nilai'.
    nim: Integer, NIM mahasiswa.

  Yields:
    String: Informasi nilai mata kuliah.
  """
  for i, item in enumerate(data_nilai):
    if (nim % 2 != 0 and i % 2 == 0) or (nim % 2 == 0 and i % 2 != 0): 
      yield f"Nilai {item['matkul']}: {item['nilai']}"

# Cek nilai awal
print("Nilai awal:", nilai)

# Menambahkan nilai ke dalam daftar
nilai_update = tambah_nilai(nilai, 'Strukdat', 15)  # Menggunakan 'Strukdat' agar nilai bisa diupdate

# Cek nilai setelah ditambahkan
print("Nilai update:", nilai_update)

# Mencetak nilai dengan generator
nim_ganjil = 202210370311025  # Ganti dengan NIM ganjil
nim_genap = 202210370311026   # Ganti dengan NIM genap

print(f"\nCetak nilai untuk NIM {nim_ganjil}:")
for nilai_cetak in cetak_nilai(nilai_update, nim_ganjil):
  print(nilai_cetak)

print(f"\nCetak nilai untuk NIM {nim_genap}:")
generator_nilai = cetak_nilai(nilai_update, nim_genap)
print(next(generator_nilai))
```
