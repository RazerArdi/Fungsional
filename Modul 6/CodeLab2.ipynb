```python
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw, ImageFont, ImageFilter

# Membuka gambar kucing
img = Image.open('kucing.jpg')

# Menambahkan teks dan anotasi langsung pada gambar menggunakan PIL
draw = ImageDraw.Draw(img)
# Anda perlu mengarahkan path ke font yang ingin digunakan
font = ImageFont.truetype('arial.ttf', 20)
draw.text((10, 10), 'Teks pada Gambar', fill='black', font=font)
draw.text((10, 50), 'Anotasi', fill='black', font=font)

# Menyimpan gambar sementara untuk rotasi dan filter
img.save('temp_image.png')

# Membuka dan memproses gambar dengan PIL
temp_img = Image.open('temp_image.png')

# Rotasi gambar
rotated_img = temp_img.rotate(45)

# Terapkan filter kontur
filtered_img = rotated_img.filter(ImageFilter.CONTOUR)

# Menyimpan hasil akhir
filtered_img.save('final_image.png')

# Menampilkan kedua gambar berdampingan menggunakan Matplotlib
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Menampilkan gambar sementara (temp_img)
axes[0].imshow(temp_img)
axes[0].set_title('Gambar Kucing dengan Teks dan Anotasi')
axes[0].axis('off')

# Menampilkan gambar hasil akhir (filtered_img)
axes[1].imshow(filtered_img)
axes[1].set_title('Gambar Kucing setelah Rotasi dan Filter')
axes[1].axis('off')

# Menampilkan plot
plt.show()

# Menampilkan hasil akhir menggunakan PIL
filtered_img.show()
```
