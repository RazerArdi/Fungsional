```python
from PIL import Image, ImageEnhance
import matplotlib.pyplot as plt
```

```python
# Baca gambar
img = Image.open('UMM.webp')
```

```python
# Peningkatan saturasi
enhancer = ImageEnhance.Color(img)
img_baru = enhancer.enhance(2.0)  # Menambahkan saturasi (nilai 2.0 adalah contoh, bisa disesuaikan)
```

```python
# Tentukan area yang ingin diperbesar
x, y, w, h = 100, 100, 200, 200  # Mengatur area zoom (diubah sesuai kebutuhan)
zoomed_img = img_baru.crop((x, y, x + w, y + h))
```

```python
# Konversi ke RGB untuk Matplotlib
img_rgb = img.convert('RGB')
img_baru_rgb = img_baru.convert('RGB')
zoomed_img_rgb = zoomed_img.convert('RGB')
```

```python
# Plot gambar sebelum dan sesudah dengan zoom
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(img_rgb)
plt.title('Gambar Awal')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(img_baru_rgb)
plt.title('Gambar Setelah Diproses')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(zoomed_img_rgb)
plt.title('Area Zoom')
plt.axis('off')

plt.show()
```
